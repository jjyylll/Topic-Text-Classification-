```python
text_document = """ We are Abron and Joyal, students of BSc Digital Business and Data Science at the University of Europe.
In our course, we are learning not only about the technology field but also about various business and management concepts.
During our first semester, we started learning programming using Python.Python is a widely used programming language in web applications, software development, data science, and machine learning.
Throughout the course, we have completed several Python-based projects.
Our final evaluation project focused on topic classification of text documents.
These projects and practical examples helped us gain a better understanding of programming and motivated us to explore the world of technology."""

with open ("text_document.txt","w") as file :    # opening  the file
        file.write(text_document)
print("text_document.txt created")                # print file name
```

```python
#importing string to access puncuations
import string

# creating class for document classification
class DocumentClassifier:

  def __init__(self,topic_keywords):
    self.topic_keywords = topic_keywords

#reading the text from a file

  def read_file(self,filename):
    try:
      with open (filename,"r") as file:      #opening the file
        text = file.read()
      if text.strip() == "":
       return None
      else:
        return text                       #checking file empty or not
    except FileNotFoundError:
      print(f"file {filename} not found")
    return None

    #cleaning the text
  def clean_text(self,text):
    text =text.lower()
    for ch in string.punctuation:
        text = text.replace(ch,"")
    words = text.split()
    return words

#counting words
  def count_words(self,word_list):
     word_count ={}
     for word in word_list:
      if word in word_count:
        word_count[word]+=1
      else:
        word_count[word]=1
     return word_count

#topic score
  def topic_score(self,word_count):
    topic_score ={}
    for topic in self.topic_keywords:
      topic_score[topic] =0                       #loop through each topic
      for keyword in self.topic_keywords[topic]:
        if keyword in word_count:
          topic_score[topic]+=word_count[keyword]
    return topic_score

#predicting topic
  def predict_score(self,topic_score):
    if not topic_score or all(score == 0 for score in topic_score.values()):
      return None
    max_topic = max(topic_score,key = topic_score.get)
    return max_topic

# method to display result

  def display_result(self,word_count,topic_scores,predicted_topic):
    print("\nTop words:")

    def get_count(item):
      return item[1]                        # sort word by frequeny

    sorted_words = sorted(word_count.items(),key = get_count,reverse = True)
    for word,count in sorted_words[:10]:
      print(f"{word}: {count}")

    print("\nTopic scores:")
    for topic,score in topic_scores.items():
      print(f"{topic}: {score}")


    print("\nPredicted topic:")
    if predicted_topic is not None:
      print(predicted_topic)
    else:
      print("No topic predicted")

# major topic keywords
  def main(self):
    topic_keywords = {
        "technology":["python","software","development","system","application","web","machine learning"],
         "social":["climate","inflation","globalization","population","environment"],
        "studies":["experience","education","knowledge"]

    }

    classifier = DocumentClassifier(topic_keywords)

    filename = input("Enter the file name: ")    #asking for file name
    text = classifier.read_file(filename)

    if text is None:
      print("Empty file ")
    else:

      words = classifier.clean_text(text)
      word_count = classifier.count_words(words)
      topic_scores = classifier.topic_score(word_count)
      predicted_topic = classifier.predict_score(topic_scores)
      classifier.display_result(word_count,topic_scores,predicted_topic)

# program entry point
if __name__ == '__main__':
# create initial object and run
    initial_classifier_instance = DocumentClassifier(topic_keywords={})
    initial_classifier_instance.main()
```
